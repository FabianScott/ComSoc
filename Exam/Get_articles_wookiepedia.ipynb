```python
import requests
import numpy as np
import pandas as pd
from tqdm import tqdm
import re
```

```python
BASE_URL = "https://starwars.fandom.com/api.php"

names = []
with open('Character_names.txt', 'r') as file:
    for name in file.readlines():
        # Ignore '\n'
        names.append(name[:-1])
```

```python
jsons = []
S = requests.Session()

for name in tqdm(names):
    URL = BASE_URL
    PARAMS = {
    "action": "parse",
    "page": name,
    "format": "json"
}

    R = S.get(url=URL, params=PARAMS)
    try:
        DATA = R.json()
        jsons.append(DATA)
    except:
        print('Something failed')
```

```python
l = []
n_errors = 0
for json in tqdm(jsons):
    try:
        text = json['parse']['text']
        title = json['parse']['title']
        l.append([title, text])
    except KeyError:
        n_errors += 1
        pass

df = pd.DataFrame(l)
print(n_errors)
```

```python
df.to_csv('Text_dump.csv')
```

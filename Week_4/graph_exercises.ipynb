```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2, 3])
G.add_nodes_from([
    (4, {'color': 'red'}),
    (5, {'color': 'green'})
])
H = nx.path_graph(10)
# Either add all nodes from another network
G.add_nodes_from(H)
# Or add the network iself as a node (or both as here)
G.add_node(H)
```

```python
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple
G.add_edges_from([(1, 2), (1, 3)])
# Copy over as before
G.add_edges_from(H.edges)
G.clear()
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
print(G.number_of_nodes())
print(G.number_of_edges())
```

```python
DG = nx.DiGraph()
DG.add_edge(2, 1)   # adds the nodes in order 2, 1
DG.add_edge(1, 3)
DG.add_edge(2, 4)
DG.add_edge(1, 2)
assert list(DG.successors(2)) == [1, 4]
assert list(DG.edges) == [(2, 1), (2, 4), (1, 3), (1, 2)]
nx.draw(DG)
```

```python
# Inspecting the graph:
print(list(G.nodes))
print(G.edges)
print(G.adj[1], G.neighbors(1))
print(G.degree[1])  # number of edges into '1'
# Can also be done with an nbunch, meaning an iterable of nodes
print(G.edges([1,2]))
# Or all:
print(G.edges(None))
```

```python
# Defining a network that resembles a square with networkx

A = nx.Graph()
A.add_nodes_from([1, 2, 3, 4])
A.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
A.add_edges_from([(2,4)])

nx.draw(A)
```

```python
B = nx.Graph()
B.add_nodes_from([1, 2, 3, 4])
B.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
B.add_edges_from([(2,4), (1,3)])

nx.draw(B)
```

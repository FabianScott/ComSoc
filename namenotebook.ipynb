```python
from bs4 import BeautifulSoup ##A package to work with HTML data
import requests #A package to make HTTP requests
import re
import numpy as np
```

```python
LINK = "https://2019.ic2s2.org/posters/"
r = requests.get(LINK) 
soup = BeautifulSoup(r.content)

div = soup.find("div",{"class":"col-md-8 page-content-wrap"})
rawstring_list = str(div).split("\n")[5:-2]

name_list = []
for line in rawstring_list:
    if line.startswith("<li>"):
        line = line[4:]
    
    if line.startswith("<"):
        continue
    
    idx = line.find("<")
    if idx:
        line = line[:idx]
    
    line_list = re.split(", | and ", line)
    
    for name in line_list:
        name_list.append(name)

len(np.unique(name_list))
```

```python
import camelot

tables = camelot.read_pdf('2021names.pdf', pages="6-47")
```

```python
dfs = [table.df for table in tables]
set1 = set()

for df in dfs[:6]:
    table = list(df[1][1:])
    for names in table:
        names = names.replace("\n", " ")
        [set1.add(i) for i in re.split(", | and ", names)]

set1
```

```python
import PyPDF2

with open("2021names.pdf", "rb") as file:
    page5 = PyPDF2.PdfReader(file).pages[5]
    page5 = str(page5.extract_text())

lines = page5.split("\n")
for line in lines:
    words = line.split(" ")
    for i, w in enumerate(words):
        if w == "":
            words[i] = " "
        
        print(words)
```

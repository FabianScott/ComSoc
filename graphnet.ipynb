```python
import networkx as nx
G = nx.Graph()

G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

H = nx.path_graph(10)
G.add_nodes_from(H)
G.add_node(H)
```

```python
G.number_of_edges(), G.number_of_nodes()
```

```python
list(G.nodes())
```

```python
# Defining a network that resembles a square with networkx

A = nx.Graph()
A.add_nodes_from([1, 2, 3, 4])
A.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
A.add_edges_from([(2,4)])

nx.draw(A)
```

```python
B = nx.Graph()
B.add_nodes_from([1, 2, 3, 4])
B.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
B.add_edges_from([(2,4), (1,3)])

nx.draw(B)
```

```python
C = nx.Graph()
C.add_nodes_from([1, 2, 3, 4, 5, 6])
C.add_edges_from([(1, 2), (2, 3), (3, 1)])
C.add_edges_from([(4, 5), (5, 6), (6, 4)])

nx.draw(C)
```
